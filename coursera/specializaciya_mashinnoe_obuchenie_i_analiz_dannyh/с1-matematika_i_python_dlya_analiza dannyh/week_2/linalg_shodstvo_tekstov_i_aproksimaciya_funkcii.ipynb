# Задача 1: сравнение предложений

```python
!cat sentences.txt
```

```python
with open('sentences.txt', 'r') as s:
    li = s.readlines()
```

```python
li = [list(filter(None, re.split('[^a-z]', i.lower()))) for i in li]
```

```python
x = set()
for i in li:
    for j in i:
        x.add(j)
```

```python
df = np.zeros((22, 254))
```

```python
for i, j in enumerate(li):
    for y, z in enumerate(x):
        if z in j:
            df[i, y] += 1
```

```python
df = pd.DataFrame(df, columns=x)
```

```python
df
```

```python
result =  [(i, cosine(df.loc[0], df.loc[i])) for i in range(22)]
```

```python
result.sort(key=lambda x: x[1])
```

```python
result = [i[0] for i in result[1:3]]
```

```python
with open('submission-1.txt', 'w') as s1:
    s1.write(str(' '.join(str(i) for i in result)))
```

# Задача 2: аппроксимация функции

Рассмотрим сложную математическую функцию на отрезке [1, 15]:

f(x) = sin(x / 5) * exp(x / 10) + 5 * exp(-x / 2)

```python
def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
```

```python
x = range(1, 16)
y = [f(i) for i in x]
```

```python
plt.plot(x, y)
```

```python
x1 = np.array([1, 15])
x11 = np.array([1, 1**1])
x12 = np.array([1, 15**1])
y1 = np.array([f(i) for i in x1])
```

```python
solve((x11, x12), y1)
```

```python
x2 = np.array([1, 8, 15])
x21 = np.array([1, 1**1, 1**2])
x22 = np.array([1, 8**1, 8**2])
x23 = np.array([1, 15**1, 15**2])
y2 = np.array([f(i) for i in x2])
```

```python
solve((x21, x22, x23), y2)
```

```python
x3 = np.array([1, 4, 10, 15])
x31 = np.array([1, 1**1, 1**2, 1**3])
x32 = np.array([1, 4**1, 4**2, 4**3])
x33 = np.array([1, 10**1, 10**2, 10**3])
x34 = np.array([1, 15**1, 15**2, 15**3])
y3 = np.array([f(i) for i in x3])
```

```python
solve((x31, x32, x33, x34), y3).round(2)
```

```python
plt.plot(x, y)
plt.plot(x1, y1)
plt.plot(x2, y2)
plt.plot(x3, y3)
plt.legend(['x', 'x1', 'x2', 'x3'])
```

```python
with open('submission-2.txt', 'w') as s2:
    s2.write(str(' '.join(str(i) for i in solve((x31, x32, x33, x34), y3).round(2))))
```
